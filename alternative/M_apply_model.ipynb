```python
import os
import sys
import numpy as np
import torch as pt
import matplotlib.pyplot as plt
from glob import glob
from tqdm import tqdm
```

```python
class Dataset(pt.utils.data.Dataset):
    def __init__(self, X_l):
        super(Dataset, self).__init__()
        self.X_l = [pt.from_numpy(x).reshape(-1,2) for x in X_l]

    def __len__(self):
        return len(self.X_l)

    def __getitem__(self, k):
        return self.X_l[k].T.unsqueeze(0).float()
```

```python
# parameters
save_path = "model/save/b_v0_0_2022-10-20_09-08"
#save_path = "model/save/b_v0_1_2022-10-20_09-29"
```

```python
# add module to path
if save_path not in sys.path:
    sys.path.insert(0, save_path)
    
# load functions
from config import config_data, config_model
from model import Model
```

```python
# locate runtime data
data_filepaths = glob("runtime_data/*.npz")

# device
device = pt.device("cuda")

# create model
model = Model(config_model).to(device)

# reload model
model.load_state_dict(pt.load(os.path.join(save_path, "model.pt"), map_location=pt.device("cpu")))

# set model to inference
model = model.eval().to(device)
```

```python
# labels
t = ["bg", "pS"]

# for all dataset
results = {}
for data_filepath in data_filepaths:
    # load data and create dataset
    f = np.load(data_filepath, allow_pickle=True)
    dataset = Dataset(list(f['X']))

    # apply model on all events
    P = []
    with pt.no_grad():
        for i in tqdm(range(len(dataset))):
            # fetch data
            X = dataset[i]

            # forward propagation
            z = model.forward(X.to(device))
            p = pt.softmax(z.detach(), axis=1)

            P.append(p.cpu().numpy())

    # create array
    P = np.concatenate(P)
    results[os.path.basename(data_filepath)[:-4]] = P

    # statistics
    mp = np.mean(np.round(P), axis=0)
    sp = np.std(np.round(P), axis=0)

    # plot
    plt.figure(figsize=(4,3))
    plt.bar(np.arange(mp.shape[0]), mp)
    for i in range(len(mp)):
        plt.text(i, 0.1, "{:.1f}% ± {:.1f}%".format(mp[i]*1e2, sp[i]*1e2), horizontalalignment='center')
    plt.xticks(np.arange(mp.shape[0]), t)
    plt.title(os.path.basename(data_filepath)[:-4])
    plt.ylim(0.0, 1.0)
    plt.tight_layout()
    plt.savefig("graphs/{}.png".format(os.path.basename(data_filepath)[:-4]), dpi=200)
    plt.show()

    print(os.path.basename(data_filepath)[:-4])
    for i in range(len(mp)):
        print("{}: {:.1f}% ± {:.1f}%".format(t[i], mp[i]*1e2, sp[i]*1e2))
```
