```python
metadata_naming = {
    'date': ['', '000000', ' '], #Match in filepath -> OK
    'recording': ['R', 'nan', ' '], #Match in name -> OK
    'pore_quantity': ['Pc_', 'nan', ' '], #Match in name -> ToDo (regex)
    'analyte_quantity': ['Ac_', 'nan', ' '], #Match in name -> ToDo (regex)
    'pore': ['P_', 'K238A', ' '], #Just add it no problem -> OK
    'analyte': ['A_', 'nan', ' '], #Match in filepath -> ToDo (rename folder to A_P1 or A_P2)
    'buffer': ['B_', '1MKCl10mMHEPESpH7p8', ' '],
    'buffer_quantity': ['Bc_', '190uL', ' '],#Always same, just add it -> OK
    'temperature': ['T_', '25', ' '], # Always same, just add it -> OK
    'voltage': ['V_', 'nan', ' '], # From folder/path -> ToDo (Separate in folders the values, folder name V_100 V_130 ...)
    'channel': ['_CH', 'nan', ''], # From filename -> OK
    'count': ['_', 'nan', '.abf'], # From filename -> OK
}

metadata_rules_here = {
    'pore': [r'P_(.*?) ', '', str],
    'pore_quantity': [r'Pc_(.*?) ', '', str],
    'analyte': [r'A_(.*?) ', '', str],
    'analyte_quantity': [r'Ac_(.*?) ', '', str],
    'temperature': [r'T_(.*?) ', -1, int],
    'voltage': [r'V_(.*?) ', -1, int],
    'channel': [r'_CH([0-9]+)', '', str],
    'count': [r'_CH[0-9]{3}_([0-9]{3})', '', str],
    'date': [r'/([0-9]{6})', '', str],
    'buffer': [r'XXXX', '1MKCl10mMHEPESpH7p8', str],
    'buffer_quantity': [r'XXXX', '190uL', str],
    'measure': [r'(.*)', '', str],
    'recording': [r'R([0-9]{2})', '', str], #add R_NN or RNN
}

def buildnewname(mn):
    global id_f
    keys = mn.keys()
    n = ''
    for key in keys:
        if key == 'channel':
            n = n + 'f' + str(id_f)
            # n = n + 'f' + str(uuid.uuid1())
            id_f = id_f + 1
            
        n = n + str(mn[key][0]) + str(mn[key][1]) + str(mn[key][2])
    return n

def getmetatatanaming(fp):
    mn = copy.deepcopy(metadata_naming)
    md = extract_metadata(fp, metadata_rules_here)
    
    if(re.search(r'[0-9]', md['analyte_quantity'])):
        if re.search(r'p[0-9]', md['analyte_quantity']):
            pass
        else:
            md['analyte_quantity'] = md['analyte_quantity'] + "p0"
    
    if(re.search(r'[0-9][0-9]', md['pore_quantity'])):
        if re.search(r'p[0-9]', md['analyte_quantity']):
            pass
        else:
            md['pore_quantity'] = md['pore_quantity'][0] + "p" + md['pore_quantity'][1]
    
    if(re.search(r'P[0-9]_', md['analyte'])):
        pass
    else:
        md['analyte'] = "MP446_" + md['analyte']
    
    if md['analyte'] == 'MP446_P1':
        md['analyte_quantity'] = '22p05uM'
        
    if md['analyte'] == 'MP446_P2':
        md['analyte_quantity'] = '20p77uM'
        
    if md['analyte'] == 'MP446_P1_Cy5':
        md['analyte_quantity'] = '25p64uM'
        
    if md['analyte'] == 'MP446_P2_Cy5':
        md['analyte_quantity'] = '25p64uM'
        
    md['pore_quantity'] = md['pore_quantity'] + 'uL'
        
        
    #IMPORTRANT
    for k in mn.keys():
        mn[k][1] = md[k]
        
    return(mn)
```

```python
id_f = 0
nns_fps = [[buildnewname(getmetatatanaming(fp)), fp]  for fp in glob("raw_data/**/*.abf", recursive=True)]
```
