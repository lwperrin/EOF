```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.cluster import KMeans
from sklearn.neighbors import KernelDensity
from scipy.optimize import curve_fit
from scipy.special import erf

import transport_signal_processing as tsp
from data_manager import DataManager
```

## Select and load events

```python
# parameters
path = "K238A_25_1MKCl10mMHEPESpH7p8/MP446_P3_2/51p28uM/*"
level = 1
selected_only = True
#condition_keys = ['analyte', 'voltage', 'date', 'recording', 'channel', 'count']
condition_keys = ['analyte', 'voltage']

# setup database connector
sigman = DataManager('database')

# load segments informations
sinfo_l = sigman.load_info(path, 's*')

# convert info to dataframe
df = pd.DataFrame(sinfo_l)

# add full condition key
df['condition'] = df.apply(lambda x: '-'.join([str(x[ck]) for ck in condition_keys]), axis=1)

# mininum number of events
df = df[df['N_cores']*df['ratio_sel'] > 0]

# keep only selected segments
df = df[(df['selected'] > (level-1))]
df
```

```python
df['condition'].unique()
```

```python
# parameters
key_sel = 'condition'

# load stats grouped by selected key
stats_dict = tsp.utils.load_stats_for_key(sigman, df.to_dict('records'), key_sel, selected_only=selected_only)

# find keys
key_sels = np.unique(df[key_sel].values)

events_dict = {}
for key in tqdm(key_sels):
    # subset
    dfs = df[df[key_sel] == key].copy()

    # load core events
    events_dict[key] = tsp.utils.load_core_events(sigman, dfs.to_dict('record'), selected_only=selected_only)
```

```python
def load_range_events(sigman, sinfo_l):
    # load core events
    rngs_all = []
    for i in range(len(sinfo_l)):
        # get path within the database
        sinfo = sinfo_l[i]
        curr_path = sigman.define_db_path(sinfo)

        # load cores & reduced and filter keep only selected ones
        r = sigman.load_data(curr_path, 's{}-events-range'.format(sinfo['sid']))[0]

        # append data
        rngs_all += list(r)
        
    return rngs_all
```

```python
# parameters
key_sel = 'condition'

# load stats grouped by selected key
stats_dict = tsp.utils.load_stats_for_key(sigman, df.to_dict('records'), key_sel, selected_only=selected_only)

# find keys
key_sels = np.unique(df[key_sel].values)

ranges_dict = {}
for key in tqdm(key_sels):
    # subset
    dfs = df[df[key_sel] == key].copy()

    # load core events
    ranges_dict[key] = load_range_events(sigman, dfs.to_dict('record'))
```

```python
# parameters
tlim = (5e-2, 50)
nbins = 30

keys = [
    'K238A-MP446_P1-100',
    'K238A-MP446_P1-130',
    'K238A-MP446_P1-160',
]
keys = df['condition'].unique()

plt.figure()
for key in keys:
    # extract dwell time for all events
    events = events_dict[key]
    dwt = np.array([evt[-1,0]-evt[0,0] for evt in events if len(evt) > 1])

    c, x = np.histogram(np.log10(dwt), range=(np.log10(tlim[0]), np.log10(tlim[1])), bins=nbins)
    c = c / np.sum(c)

    x = np.power(10.0, x)
    x = 0.5*(x[:-1] + x[1:])
    
    mdwt = np.power(10.0, np.mean(np.log10(dwt)))
    print("{}: {:.3f}".format(key, mdwt))

    plt.semilogx(x, c, '.-', label=key)
    
plt.xlim(tlim)
plt.legend(loc='best')
plt.ylabel('density')
plt.xlabel('dwell time [ms]')
plt.show()
```

```python
# parameters
tlim = (1e-2, 2)
nbins = 10
tlim_fit = [0.3, 1.5]

keys = [
    'K238A-MP446_P1-100',
    'K238A-MP446_P1-130',
    'K238A-MP446_P1-160',
]
keys = df['condition'].unique()

plt.figure()
for key in keys:
    # extract dwell time for all events
    events = events_dict[key]
    dwt = np.array([evt[-1,0]-evt[0,0] for evt in events if len(evt) > 1])

    c, x = np.histogram(dwt, range=(tlim[0], tlim[1]), bins=nbins)
    #c = c / np.sum(c)
    c = c / np.max(c)

    x = 0.5*(x[:-1] + x[1:])

    m = ((x >= tlim_fit[0]) & (x <= tlim_fit[1]))

    p0, p1 = np.polyfit(x[m], np.log10(c[m]), 1)

    xf = np.linspace(tlim_fit[0], tlim_fit[1], 100)
    yf = p0*xf + p1
    
    print("{}: {:.3f}".format(key, -1.0/p0))
    
    plt.plot(x,np.log10(c), label=key)
    plt.plot(xf,yf,'k--')

plt.xlim(tlim)
plt.legend(loc='best')
plt.ylabel('rescaled count')
plt.xlabel('dwell time [ms]')
plt.show()
```
